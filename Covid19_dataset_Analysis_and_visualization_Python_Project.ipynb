#### Total number of confirmed, death, recovered, active cases per WHO region

```python
data = df[['Confirmed', 'Deaths', 'Recovered', 'Active', 'WHO Region']].groupby('WHO Region').mean().round(2).sort_values(by = 'Confirmed', ascending = False)

data
```

```python
data['Confirmed'].plot(kind='bar')

plt.title('Confirmed Cases Grouped By WHO Region')

plt.ylabel('Count In Million')
```

#### Percentage of recovery for each region


```python
data['Recovery Percentage'] = (data['Recovered'] / data['Confirmed']) * 100

print(data['Recovery Percentage'])
```

```python
data
```

#### Percentage of death for each region

```python
data['Death Percentage'] = (data['Deaths'] / data['Confirmed']) * 100

data['Death Percentage']
```

```python
data
```

#### Percentage of active cases for each region

```python
data['Active Percentage'] = (data['Active'] / data['Confirmed']) * 100

data['Active Percentage']
```

```python
data
```

```python
data = data.reset_index()

data
```

```python
plt.pie(data['Recovered'], labels = data['WHO Region'], autopct = '%0.2f%%', radius = 1.05)

plt.title("Percentage of Recovery in each Region")

plt.show()
```

#### Which Country experiences the Maximum Frequency of death cases

```python
max_freq = df[['Country/Region', 'Deaths / 100 Cases']].sort_values(by = 'Deaths / 100 Cases', ascending = False).head(10)

max_freq
```

```python
plt.figure(figsize = (10, 7))

sns.barplot(x = 'Country/Region', y = 'Deaths / 100 Cases', data = max_freq)

plt.xticks(rotation = 35)

plt.show()
```

##### Maximum frequency of death cases are occured in yemen

#### Which country experiences the max frequency of recovery case

```python
recover_freq = df[['Country/Region', 'Recovered / 100 Cases']].head(10).sort_values(by = 'Recovered / 100 Cases', ascending = False)

recover_freq 
```

```python
plt.figure(figsize = (10,7))

sns.barplot(x = 'Country/Region', y = 'Recovered / 100 Cases', data = recover_freq)

plt.xticks(rotation = 25)

plt.show()
```

##### The maximum frequency of recovered cases are mainly in 2 countries such as, Austria and Andorra.

#### Which country rises in new cases

```python
new_cases = df[['Country/Region', 'New cases']].sort_values(by = 'New cases', ascending = False)[0:10]

new_cases
```

```python
plt.figure(figsize = (10, 7))

sns.barplot(x = 'Country/Region', y = 'New cases', data = new_cases)

plt.xticks(rotation = 25)

plt.show()
```

#### Most of the new cases are rises in US
